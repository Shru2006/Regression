#Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
print(dataset)
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

#Training the linear regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y) #since the dataset is small it is not split to training & test set
```

#Training the Polynomial Regression model on whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2) #based on how accurate or close you want your predicted values
x_poly = poly_reg.fit_transform(x) #to give the poly effect/power to x
lin_reg_2 = LinearRegression() 
lin_reg_2.fit(x_poly, y) #to create linear regression for polynomial
```

#Visualising the Linear Regression results

```python
plt.scatter(x, y)
plt.plot(x, lin_reg.predict(x), color = 'blue') #real salary & predicted salary using linear regression model
plt.title('Truth/Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

#Visualising the Polynomial Regression results

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg_2.predict(poly_reg.fit_transform(x)), color = 'blue') #real salary & predicted salary using polynomial regression model
plt.title('Truth/Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

#Visualising the Polynomial Regression results (for higher resolution & smoother curve)

```python
x_grid = np.arange(min(x), max(x), 0.1) #to divide axis into smaller chunks separated by 0.1 distance
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color = 'blue') #real salary & predicted salary using polynomial regression model
plt.title('Truth/Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

#Predicting a new result with Linear Regression

```python
lin_reg.predict([[6.5]]) #[] is for List/Vector; [[]] is for 2-D Arrays
```

#Predicting a new result with Polynomial Regression

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]])) #gives an output which is closer to real value of salary
```
